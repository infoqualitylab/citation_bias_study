```python
import igraph
from igraph import * #importing everything in igraph
import igraph as ig
import pandas  #dataframe library
import numpy  #numeric python, basically gives you matrice
```

```python
HTTLPR = pandas.read_csv("HTTLPR.csv") #reading the .csv file (excel spreadsheet)
new = pandas.DataFrame(numpy.zeros(shape=(73, 73)), columns=HTTLPR['Study'], index=HTTLPR['Study']) #making a data frame
```

```python
randyears = input("What is the year gap?")
randyears = int(randyears)
```

```python
for a, ayear in zip(HTTLPR['Study'], HTTLPR['Year']):
    for b, byear in zip(HTTLPR['Study'], HTTLPR['Year']):  #pairs the two lists together, converting the dataif a == b:
        continue
        if ayear >= byear + randyears:citation.at[a, b] = 1 #filling the new data frame with a1 if they have an X
```

```python
g = igraph.Graph.Adjacency((new.values > 0).tolist())  #new.values is a numpy array stored in the data frame
```

```python
g.vs["label"] = new.index
g.vs["Year"] = HTTLPR['Year']
g.vs["YearOnline"] = HTTLPR['YearOnline']
g.vs["Outcome"] = HTTLPR['Outcome']
g.vs["Abstract"] = HTTLPR['Abstract'] #all of this section is just assigning the labels and the different attributes to each other
```

```python
visual_style = {}

out_name = "greenburg2.png"
```

```python
visual_style["bbox"] = (2200, 2200)

visual_style["margin"] = 27

visual_style["vertex_size"] = 50

visual_style["vertex_label_size"] = 30

visual_style["edge_curved"] = False
```

```python
num_colors = max(g.degree()) + 1
palette = ig.RainbowPalette(n=num_colors) #can change the different parameters and class but rainbow is prettiest
color_list = [palette.get(degree) for degree in g.degree()] #g is my graph
visual_style["vertex_color"] = color_list
```

```python
plot(g, out_name, **visual_style)
```
